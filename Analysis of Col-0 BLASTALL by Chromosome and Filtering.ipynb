```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot
import seaborn as sns
```

```python
path = r"C:\Users\Christie and Victori\PycharmProjects\Satellite_Research_Project\frequency_t2t\000test\t2t_col_cen180_Chr1_mostfreq.Fasta_t2t_col_cen180_Chr1dropfreq.Fasta.out"
chr1freq = pd.read_csv(path)

#blast output format:
#query id, subject id, % identity, alignment length, mismatches, gap opens, 
#q. start, q. end, s. start, s. end, evalue, bit score, sequence

ref = []
query = []
similarity = []
evalue = []

for k in range(len(chr1freq.index)):
    element = chr1freq.iloc[k,0] 
    splitted = element.split("\t") #split string into accession contig/chromosome and size of contig/chromosome
    ref.append(splitted[1]) #append name of reference sequence to list
    query.append(splitted[0])  #append name of query sequence to list
    similarity.append(float(splitted[2])) #append similarity score
    evalue.append(float(splitted[10])) # append e value
    
chr1freq["reference sequence"] = ref
chr1freq["Query sequence"] = query #add lists to columns to make dataframe
chr1freq["% similarity"] = similarity
chr1freq["e value"] = evalue
chr1freq.drop(columns=chr1freq.columns[0], axis=1, inplace=True) #remove original column
chr1freq #shows dataframe
```

```python
plt.pyplot.figure(figsize=(10,7), dpi= 80) #change size of plot

x = chr1freq["% similarity"] #plot similarity scores for chr1
kwargs = dict(hist_kws={'alpha':.3}, kde_kws={'linewidth':3}) # alpha changes opacity of bars, kde plots line on graph
sns.distplot(x, color="dodgerblue", label="Compact", **kwargs) 
plt.pyplot.title("Investigation into the similarity of repeats compared to the most frequent repeat for Chr1")
plt.pyplot.axvline(96, color='r')
plt.pyplot.savefig(r"C:\Users\Christie and Victori\Documents\christie\work uni\part II\project\cleaning_sequences\histograms\Similarity of repeats t2t chr1.png")
```

```python
#create list of sequences on each chromosome which are between 0-96% similarity
chr1bysim = chr1freq[chr1freq["% similarity"].between(0,96)]
sim = []

chr2bysim = chr2freq[chr2freq["% similarity"].between(0,96)]
chr3bysim = chr3freq[chr3freq["% similarity"].between(0,96)]
chr4bysim = chr4freq[chr4freq["% similarity"].between(0,96)]
chr5bysim = chr5freq[chr5freq["% similarity"].between(0,96)]
                     
sim = sim + chr1bysim["Query sequence"].tolist() + chr2bysim["Query sequence"].tolist() + chr3bysim["Query sequence"].tolist() + chr4bysim["Query sequence"].tolist() + chr5bysim["Query sequence"].tolist()
```

```python
path = r"C:\Users\Christie and Victori\Documents\christie\work uni\part II\project\cleaning_sequences\all.repeats.from.t2t-col.csv"
t2t_col_df = pd.read_csv(path)
t2t_col_df["names"] = t2t_col_df["fasta.name"].str.cat(t2t_col_df["start"].astype("str"), sep ="_") 
# concatenate strings in fasta.name and start column to make names file; make sure variables are strings
t2t_col_df["names"] = "t2t_col_" + t2t_col_df["names"]

#subsetting repeats into cen180 and cen160
cen180_t2t_col = t2t_col_df[t2t_col_df["length"].between(175, 180)]
cen180_t2t_col["names"] = "cen180_" + cen180_t2t_col["names"]
#cen180_t2t_col has 68309 rows total
```

```python
allname = cen180_t2t_col["names"].tolist()
wanted = [x for x in allname if x not in sim] #get list of wanted
```

```python
wantdf = cen180_t2t_col[cen180_t2t_col.names.isin(wanted)] #get rows from df matching list of names
#wantdf has 20657 rows total
```

```python
#need to add back in the most frequent repeats for each chromosome
countchr = pd.DataFrame(cen180_t2t_col["fasta.name"].value_counts())

mostfreqlist = []
for j in range(len(countchr.index)):
    chrname = countchr.index[j]
    Chr = cen180_t2t_col.get(cen180_t2t_col["fasta.name"] == chrname)
    mostfreq = pd.DataFrame(Chr["sequence"].value_counts())
    freqseq = mostfreq.index[0]
    myrow = Chr.get(Chr["sequence"] == freqseq).iloc[0]
    mostfreqlist.append(myrow)

for i in range(len(mostfreqlist)):
    a_series = pd.Series(mostfreqlist[i], index = wantdf.columns)
    wantdf = wantdf.append(a_series, ignore_index=True)
#wantdf has 20667 rows
```

```python
wantdf = wantdf.drop_duplicates(subset = ["sequence"])
#wantdf now has 9680 rows
```

```python
#make UNIQUE cen 180 fasta

uniq_fasta = open(r"t2t_col_blastfiltered_uniquecen180.Fasta","w+") #open fasta file, "w+" means you intend to write in it
empty_uniq_fasta = [] #empty list to compile all rows
for j in range(len(wantdf)): #do for all rows
    string_uniq = wantdf.iloc[j,4]
    n=60
    split_uniq = [string_uniq[i:i+n] for i in range(0, len(string_uniq), n)] #splits sequence string into length 60
    join_uniq = "\n".join(split_uniq) #joins string with new lines
    write_fasta_uniq = ">"+wantdf.iloc[j,8]+"\n"+join_uniq #create fasta format
    empty_uniq_fasta.append(write_fasta_uniq) #append row's fasta to list
joined_fasta_uniq = "\n".join(empty_uniq_fasta) #join strings in list into one big string with new line breaks
uniq_fasta.write(joined_fasta_uniq) #enter list into fasta file
uniq_fasta.close() #close fasta file
```
